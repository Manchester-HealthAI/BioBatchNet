# macaque

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
sns.set_context("talk", font_scale=1.2)

file_path = "Tables_Organized.xlsx"
sheet_name = "IMMU"     
df = pd.read_excel(file_path, sheet_name=sheet_name) 

batch_effect_removal = df[["ASW Batch", "Graph Conn.", "ILISI", "PCR"]].mean(axis=1)
bio_conservation = df[["ASW Cell", "NMI", "ARI"]].mean(axis=1)
models = df["Model"]

# Custom colors: highlight "my_model" with a unique color
base_colors = sns.color_palette("Set2", n_colors=len(models))
highlight_color = 'darkred'  # Choose a highlight color for "my_model"
colors = [highlight_color if model == "BERD" else base_colors[i] for i, model in enumerate(models)]

# Batch Effect Removal
plt.figure(figsize=(9, 6))
ax1 = sns.barplot(x=models, y=batch_effect_removal, palette=colors)
plt.grid(False)
plt.ylabel("Batch Effect Removal Average Scores", fontsize=14)
plt.ylim(0, 1.1)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)

ax1.set_xlabel("")
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.spines['left'].set_color('black')
ax1.spines['bottom'].set_color('black')
ax1.spines["left"].set_linewidth(1.5)
ax1.spines["bottom"].set_linewidth(1.5)


for index, value in enumerate(batch_effect_removal):
    plt.text(index, value + 0.02, f"{value:.2f}", ha='center', va='bottom', fontsize=12)
for tick in ax1.get_xticklabels():
    if tick.get_text() == "BERD":
        tick.set_fontweight('bold')
for bar in ax1.patches:
    current_width = bar.get_width()
    desired_width = 0.6  
    new_x = bar.get_x() + (current_width - desired_width) / 2
    bar.set_width(desired_width)
    bar.set_x(new_x)
    
plt.tight_layout()
plt.show()

# Bio Conservation
plt.figure(figsize=(9, 6))
ax2 = sns.barplot(x=models, y=bio_conservation, palette=colors)
plt.ylabel("Bio Conservation Average Scores", fontsize=14)
plt.grid(False)
plt.ylim(0, 1.1)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)

ax2.set_xlabel("")
ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)
ax2.spines['left'].set_color('black')
ax2.spines['bottom'].set_color('black')
ax2.spines["left"].set_linewidth(1.5)
ax2.spines["bottom"].set_linewidth(1.5)


for index, value in enumerate(bio_conservation):
    plt.text(index, value + 0.02, f"{value:.2f}", ha='center', va='bottom', fontsize=12)

for tick in ax2.get_xticklabels():
    if tick.get_text() == "BERD":
        tick.set_fontweight('bold')

for bar in ax2.patches:
    current_width = bar.get_width()
    desired_width = 0.6  
    new_x = bar.get_x() + (current_width - desired_width) / 2
    bar.set_width(desired_width)
    bar.set_x(new_x)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set(style="white", context="talk")
file_path = "Results.xlsx"  
sheet_name = "macaque" 
df = pd.read_excel(file_path, sheet_name=sheet_name)

my_model_name = "my_model" 

unique_models = df['Model'].unique()

other_models = [model for model in unique_models if model != my_model_name]
other_colors = sns.color_palette("Set2", n_colors=len(other_models))
color_mapping = {model: color for model, color in zip(other_models, other_colors)}
color_mapping[my_model_name] = 'darkred'  #  matplotlib 

plt.figure(figsize=(7, 7))
ax = sns.scatterplot(
    data=df, 
    x='NMI', 
    y='ARI', 
    hue='Model', 
    palette=color_mapping, 
    s=200, 
    edgecolor='black', 
    linewidth=0.8
)

plt.xlabel("NMI", fontsize=14)
plt.ylabel("ARI", fontsize=14)

plt.xlim(0.6, 1.0)
plt.ylim(0.4, 1.0)
plt.xticks(np.arange(0.6, 1.05, 0.1), fontsize=12)  #  0.1 
plt.yticks(fontsize=12)

ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_linewidth(1.5)
ax.spines["bottom"].set_linewidth(1.5)

ax.grid(False)

plt.legend(title="Model", fontsize=12, title_fontsize=12, loc='lower right', frameon=False)
plt.tight_layout()

plt.show()
```

# mouse brain

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
sns.set_context("talk", font_scale=1.2)

# Load data
file_path = "Performance_Metrics.xlsx"
sheet_name = "mouse_brain"
df = pd.read_excel(file_path, sheet_name=sheet_name)

batch_effect_removal = df[["ASW Batch", "Graph Conn.", "ILISI", "PCR"]].mean(axis=1)
bio_conservation = df[["ASW Cell", "NMI", "ARI"]].mean(axis=1)
models = df["Model"]

# Custom colors: highlight "BERD" with dark red
base_colors = sns.color_palette("Set2", n_colors=len(models))
highlight_color = 'darkred'

# Batch Effect Removal
plt.figure(figsize=(9, 6))
ax1 = sns.barplot(x=models, y=batch_effect_removal, palette=base_colors)
plt.grid(False)
plt.ylabel("Batch Effect Removal Average Scores", fontsize=14)
plt.ylim(0, 1.1)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)

ax1.set_xlabel("")
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.spines['left'].set_color('black')
ax1.spines['bottom'].set_color('black')
ax1.spines["left"].set_linewidth(1.5)
ax1.spines["bottom"].set_linewidth(1.5)

# Highlight BERD and resize bars
for index, value in enumerate(batch_effect_removal):
    plt.text(index, value + 0.02, f"{value:.2f}", ha='center', va='bottom', fontsize=12)
    if models[index] == "BERD":
        ax1.patches[index].set_facecolor(highlight_color)

for bar in ax1.patches:
    current_width = bar.get_width()
    desired_width = 0.6
    new_x = bar.get_x() + (current_width - desired_width) / 2
    bar.set_width(desired_width)
    bar.set_x(new_x)

# Bold BERD label
for tick in ax1.get_xticklabels():
    if tick.get_text() == "BERD":
        tick.set_fontweight('bold')

plt.tight_layout()
plt.show()

# Bio Conservation
plt.figure(figsize=(9, 6))
ax2 = sns.barplot(x=models, y=bio_conservation, palette=base_colors)
plt.ylabel("Bio Conservation Average Scores", fontsize=14)
plt.grid(False)
plt.ylim(0, 1.1)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)

ax2.set_xlabel("")
ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)
ax2.spines['left'].set_color('black')
ax2.spines['bottom'].set_color('black')
ax2.spines["left"].set_linewidth(1.5)
ax2.spines["bottom"].set_linewidth(1.5)

# Highlight BERD and resize bars
for index, value in enumerate(bio_conservation):
    plt.text(index, value + 0.02, f"{value:.2f}", ha='center', va='bottom', fontsize=12)
    if models[index] == "BERD":
        ax2.patches[index].set_facecolor(highlight_color)

for bar in ax2.patches:
    current_width = bar.get_width()
    desired_width = 0.6
    new_x = bar.get_x() + (current_width - desired_width) / 2
    bar.set_width(desired_width)
    bar.set_x(new_x)

# Bold BERD label
for tick in ax2.get_xticklabels():
    if tick.get_text() == "BERD":
        tick.set_fontweight('bold')

plt.tight_layout()
plt.show()
```
